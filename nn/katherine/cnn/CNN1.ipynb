Encode sequences

```python
def encode(infile):
    # For one-hot encoding
    d = {'A':[1,0,0,0],
        'C':[0,1,0,0],
        'G':[0,0,1,0],
        'T':[0,0,0,1]}
    f = open(infile, "r")
    X = []
    while True:
        line = f.readline()
        line = line.strip() # strip newline character
        if len(line) == 0:
            break
        seq = []
        for nt in line:
            seq.append(d[nt])
        X.append(seq)
    X = np.array(X)
    return X
```

```python
def get_data(file1, file0, test_size=0.33):
    # One-hot encoding
    f1 = encode(file1)
    f0 = encode(file0)
    # Concatenate f1 and f0
    X = np.concatenate((f1, f0), axis=0)
    Y = np.concatenate((np.ones(len(f0)), np.zeros(len(f0))), axis=None)  # labels
    #dataset = np.concatenate((X,Y), axis=1)
    # Shuffle
    dataset = tf.data.Dataset.from_tensor_slices((X,Y))
    dataset.shuffle(buffer_size=20000)
    # Train/test splitting
    y = [y for x,y in dataset]
    #y = np.concatenate([y for x, y in dataset], axis=0)
    return y
```
